**SESSION 1: Date & Time Operations**

*Why Date & Time Is CRITICAL in Data Analytics*

Almost every real business question is time-based:

- ðŸ“ˆ Monthly sales trend
- ðŸ“Š Year-over-year growth
- ðŸ›’ Orders in last 30 days
- ðŸ‘¥ Active users per week

```python
import pandas as pd
```

**Step 1: Create a Sample Dataset**

```python

data = {
    'order_id': [101, 102, 103, 104],
    'order_date': ['2024-01-05', '2024-02-10', '2024-02-15', '2024-03-01'],
    'sales': [2500, 1800, 3200, 1500]
}

df = pd.DataFrame(data)
df
```

**Step 2: Check the Data Type**

```python
df.info()
```

**Step 3: Convert String â†’ Datetime**

```python
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
df.info()
```

**Step 4: Extract Date Components**

```python
# Extract Year
df['year'] = df['order_date'].dt.year
```

```python
# Extract Month
df['month'] = df['order_date'].dt.month
```

```python
# Extract Day
df['day'] = df['order_date'].dt.day
```

```python
# Extract Weekday
df['weekday'] = df['order_date'].dt.day_name()
```

These columns are used directly in:

- GroupBy
- Pivot tables
- Dashboards

**Step 5: Filtering Data Using Dates**

```python
# Example 1: Orders after Feb 1, 2024

df[df['order_date'] >= '2024-02-01']
```

```python
# Example 2: Orders in February only

df[df['month'] == 2]

```

**Step 6: Date Difference**

```python
# Days since order
df['days_since_order'] = (pd.Timestamp('2024-03-10') - df['order_date']).dt.days
```

**SESSION 2: Pivot Tables**

*Why Pivot Tables Are CRITICAL for Data Analysts*

Pivot tables are used to:

- Create summary reports
- Answer business questions fast
- Replace complex GroupBy code
- Do Excel / Power BIâ€“style analysis


*What Is a Pivot Table*

A pivot table:

- Takes raw data
- Groups it
- Aggregates it
- Shows results in a clean table

GroupBy + Aggregate + Reshape


```python
import pandas as pd 
```

**Step 1: Create a Sample Dataset**

```python

data = {
    'city': ['Delhi', 'Delhi', 'Mumbai', 'Mumbai', 'Chennai', 'Chennai'],
    'year': [2023, 2024, 2023, 2024, 2023, 2024],
    'sales': [5000, 7000, 4000, 6500, 3000, 4800]
}

df = pd.DataFrame(data)
df
```

**Step 2: Understand the Business Question**

Example questions:

- Total sales per city?
- Year-wise sales per city?
- Average sales per city?

```python
df.groupby('city')['sales'].sum()
```

```python
df.groupby(['city', 'year'])['sales'].sum()
```

```python
df.groupby('city')['sales'].mean()
```

**Step 3: Basic Pivot Table**

```python
# Total Sales per City
pd.pivot_table(
    df,
    values='sales',
    index='city',
    aggfunc='sum'
)

# values â†’ what to calculate
# index â†’ group by
# aggfunc â†’ how to summarize
```

**Step 4: Pivot with Rows + Columns**

```python
# Year-wise Sales per City
pd.pivot_table(
    df,
    values='sales',
    index='city',
    columns='year',
    aggfunc='sum'
)
```

**Step 5: Multiple Aggregations**

```python
# Sum + Average Sales per City
pd.pivot_table(
    df,
    values='sales',
    index='city',
    aggfunc=['sum', 'mean']
)
```

**Step 6: Handle Missing Values in Pivot Tables**

By default:

- Missing combinations = NaN

```python
# Fill missing with 0
pd.pivot_table(
    df,
    values='sales',
    index='city',
    columns='year',
    aggfunc='sum',
    fill_value=0
)
```

**pivot() vs pivot_table()**

pivot()

- No aggregation
- Data must be unique

pivot_table()

- Supports aggregation
- Handles duplicates
- Used in real analytics

*Why This Session Exists*

Real analysts:

- Take raw data

```python
import pandas as pd
```

**Step 1: Create Raw Datasets**

```python
# Employee Table

employees = {
    'emp_id': [1, 2, 3, 4, 5],
    'name': ['Amit', 'Sara', 'John', 'Neha', 'Raj'],
    'dept_id': [101, 102, 101, 103, None],
    'salary': [50000, 60000, None, 45000, 52000],
    'join_date': ['2022-01-10', '2023-03-15', '2021-07-01', '2024-02-20', '2023-11-05']
}

df_emp = pd.DataFrame(employees)
```

```python
# Department Table
departments = {
    'dept_id': [101, 102, 103],
    'department': ['IT', 'HR', 'Finance']
}

df_dept = pd.DataFrame(departments)
```

**Step 2: Inspect the Data**

```python
df_emp.info()
df_emp.isnull().sum()
```

**Step 3: Data Cleaning**

```python
# Convert Join Date
df_emp['join_date'] = pd.to_datetime(df_emp['join_date'])
```

```python
# Handle Missing Salary
df_emp['salary'].fillna(df_emp['salary'].mean(), inplace=True)
```

```python
# Handle Missing Department
df_emp['dept_id'].fillna(999, inplace=True)
```

**Step 4: Feature Engineering**

```python
#V Extract Year
df_emp['join_year'] = df_emp['join_date'].dt.year
```

```python
# dentify Recent Joiners (2023+)
df_emp['recent_joiner'] = df_emp['join_year'] >= 2023
```

**Step 5: Merge Datasets**

```python
final_df = pd.merge(
    df_emp,
    df_dept,
    on='dept_id',
    how='left'
)
```

**Step 6: GroupBy Analysis**

```python
# Average Salary per Department
avg_salary = final_df.groupby('department')['salary'].mean()
```

```python
# Count of Employees per Department
emp_count = final_df.groupby('department')['emp_id'].count()
```

**Step 7: Pivot Table (Report-Ready Output)**

```python
pd.pivot_table(
    final_df,
    values='salary',
    index='department',
    aggfunc=['mean', 'count']
)
```

**Step 8: Filter Recent Joiners**

```python
recent_employees = final_df[final_df['recent_joiner']]
```

**Step 9: Save Final Output**

```python
final_df.to_csv('clean_employee_data.csv', index=False)
```
